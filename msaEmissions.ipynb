```python
import csv 
import pandas as pd 
import scipy as sp
import numpy as np
import statsmodels.api as sm # 
import statsmodels.formula.api as smf #
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier

temp = pd.read_csv('emissions.csv', sep='/')
```

```python

df = temp[['check', 'state1', 'population1', 'density', 'area_land','housing', 'total_emission', 'year_built', 'commute']]
df.columns = ['msa', 'states', 'population', 'density','area_land','housing_units','emissions', 'year_built','commute']
df = df[df['population'] >= 100000]
df['pop_log'] = np.log(df['population'])
df['emi_log'] = np.log((df['emissions']*1000000))
df['emi_total'] = df['emissions']*1000000
df['CO2per'] = df['emi_total']/df['population']
df['housing_log'] = np.log(df['housing_units'])
```

```python
model_total = smf.ols('emi_total ~ population', data = df).fit()
print model_total.summary()
```

```python
df['emi_log'].mean()

fig1 = plt.figure(figsize=(18,6))

ax = fig1.add_subplot(121)
ax.scatter(df["population"], df["emi_total"], s=25, alpha = .6, color = 'blue')
ax.set_ylabel('Total Emissions (Million Metric Tonnes)')
ax.set_xlabel('Total Population (Millions)')
ax.set_title("Figure 1 \nTotal Emissions vs Total Population")
ax.plot(df['population'], model_total.fittedvalues, 'r--')
ax.text(-500000, 50000000, r'Model: $Emissions = 2.45 * Pop. + 1.6M$', fontsize=15)
```

```python
model_log = smf.ols('emi_log ~ pop_log', data = df).fit()
print model_log.summary()


srt = np.sort(model_log.resid)
plt.plot(srt)
```

```python
fig1 = plt.figure(figsize=(18,6))
ax = fig1.add_subplot(122)
ax.scatter(df["pop_log"], df["emi_log"], s=25, alpha = .6, color ='blue')
ax.set_ylabel('Log Emissions')
ax.set_xlabel('Log Population')
#ax.set_xlim(0, 18)
ax.set_title("Figure 2 \nLog Emissions vs Log Population")
plt.plot(df["pop_log"],  model_log.fittedvalues, 'r--.')
ax.text(11.5, 18, r'Model: $LN(Emissions) = 0.85 * LN(Pop) + 3.4$', fontsize=15)
 
```
