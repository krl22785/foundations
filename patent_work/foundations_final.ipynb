```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
pd.set_option('display.mpl_style', 'default')
```

### Part I - Patents and Economic Development

MSA GDP

```python
dfGDP = pd.read_csv('msa_gdp.csv', usecols= (['Fips', 'Area','2008','2009','2010','2011','2012']))
dfGDP['Fips'] = dfGDP['Fips'].astype(str)
dfGDP.dtypes
```

MSA Patents

```python
dfPatents = pd.read_csv('msa_patents.csv', usecols=(['ID Code','2001','2002','2003','2004','2005']))    
dfPatents = dfPatents.rename(columns={'ID Code': 'Fips'})
dfPatents['Fips'] = dfPatents['Fips'].astype(str)
dfPatents["Fips"] = dfPatents["Fips"].map(lambda x: str(x)[1:-2]) ## be careful with this 
```

MSA Populations

```python
## 2008 to 2012
dfPopulation = pd.read_csv('msa_population.csv')   
dfPopulation = dfPopulation.dropna()

## 2001 to 2015 comment: need this for patents per US MSA dates 
dfPopulation1 = pd.read_csv('msa_population1.csv')   
dfPopulation1 = dfPopulation1.dropna()

## merge tables 
dfPopulation = pd.merge(dfPopulation1, dfPopulation, on = 'GeoFips')
dfPopulation = dfPopulation.rename(columns = {'GeoName_x': 'GeoName',
                                              'GeoFips': 'Fips'})
del dfPopulation['GeoName_y']
```

Create Patent Intensity Data
* (Patent Count / MSA Population) * 100000

```python
dateRange = ['Fips','2001','2002','2003','2004','2005']
temp = dfPopulation[dateRange]

patentPopulation = pd.merge(temp, dfPatents, on = 'Fips')

patentPopulation["population_cum"] = patentPopulation.iloc[:,1:6].apply(np.sum, axis = 1)
patentPopulation["patent_cum"] = patentPopulation.iloc[:,6:11].apply(np.sum, axis = 1)
patentPopulation['patent_intensity'] = (patentPopulation['patent_cum'] / patentPopulation['population_cum']) * 1000000


patentIntensity = pd.DataFrame(patentPopulation[['Fips','patent_intensity']])
```

GDP per Capita
* MSA GDP / Population

```python
GDPdateRange = ['Fips','2008','2009','2010','2011','2012']
temp1 = dfPopulation[GDPdateRange]

GDPPopulation = pd.merge(temp1, dfGDP, on = 'Fips')

GDPPopulation["GDP_cum"] = GDPPopulation.iloc[:,7:12].apply(np.sum, axis = 1)
del GDPPopulation['Area']

GDPPopulation["GDP_per_capita"] = GDPPopulation["GDP_cum"] / 5

GDP_per = pd.DataFrame(GDPPopulation[['Fips', 'GDP_per_capita']])
```

```python
patents = pd.merge(patentIntensity, GDP_per, on = 'Fips')
```

Visualization 

```python
import statsmodels.api as sm # 

y = np.log(patents["GDP_per_capita"])
x = np.log(patents["patent_intensity"])
results = sm.OLS(y, sm.add_constant(x)).fit()

plt.scatter(np.log(patents["patent_intensity"]), np.log(patents["GDP_per_capita"]), s = 20.0, alpha = .75)
plt.plot(np.log(patents["patent_intensity"]), results.fittedvalues, 'r--', alpha = .75)
plt.ylabel("LN(GDP per Capita)")
plt.xlabel("LN(Patent Intensity)")
plt.title("Patents and Economic Development Relationship\n")
```
